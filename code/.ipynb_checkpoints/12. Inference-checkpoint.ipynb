```python
import gc
import os
import numpy as np
import pandas as pd
import random as rn
import seaborn as sns
import tensorflow as tf
import matplotlib as mpl
import matplotlib.pyplot as plt

from tensorflow import keras
from itertools import product
from keras import backend as K
from tensorflow.keras import layers
from tensorflow.python.client import device_lib
```

```python
np.random.seed(35)
rn.seed(35)
tf.random.set_seed(35)
print(device_lib.list_local_devices())
```

```python
file_path = "../../data/BG Ex data/"
model_path = "../../model/"
file_list = os.listdir(file_path)
```

```python
fold_info = np.load("../../data/BG_Fold_Info.npz", allow_pickle=True)

FOLD = 2

test = fold_info["test"][FOLD]
```

```python
test_file = test[0]
```

```python
test_data = np.load(file_path+test_file+".npz")
test_12x = np.concatenate(test_data["gen12"]).reshape(-1, 12, 1)
test_24x = np.concatenate(test_data["gen24"]).reshape(-1, 24, 1)
test_36x = np.concatenate(test_data["gen36"]).reshape(-1, 36, 1)
test_x = {
    "test_12x":test_12x,
    "test_24x":test_24x,
    "test_36x":test_36x
}
test_y =test_data["y12"].reshape(-1, 1)
```

```python
model_path = model_path + f'Fold{FOLD}/'
model_list = os.listdir(model_path)
```

```python
def root_mean_squared_error(y_true, y_pred):
    return K.sqrt(K.mean(K.square(y_pred - y_true)))
```

```python
pred_list = []

for model in model_list:
    params = model.split("_")
    
    MODEL_NAME = params[0]
    SEQ = params[1]
    HIDDEN = params[2]
    BATCH = params[3][:-3]    
    
    MODEL = keras.models.load_model(
        model_path + model,
        custom_objects={"root_mean_squared_error": root_mean_squared_error}
    )
    
    if SEQ=="12":
        pred = MODEL.predict(test_x["test_12x"])
        pred_list.append(pred)
    elif SEQ=="24":
        pred = MODEL.predict(test_x["test_24x"])
        pred_list.append(pred)
    else:
        pred = MODEL.predict(test_x["test_36x"])
        pred_list.append(pred)
        
pred_result = np.mean(pred_list, axis=0)
```

```python
soft = pred_result.flatten().astype(int).astype(float)
real = test_y.flatten().astype(float)
naive = np.roll(real, 6)

soft = soft[6:]
real = real[6:]
naive = naive[6:]
```

```python
print("Soft:",soft)
print("Real:",real)
print("Naive:",naive)
```

```python
fitted = soft
residual = real - fitted

res_df = pd.DataFrame({
    'fitted':fitted,
    'residual':residual
})

sns.regplot(x="fitted", y="residual", lowess=True, data=res_df, line_kws={'color':'red'})
plt.plot([50, 300], [0,0], '--', color='grey')
```

```python
import scipy.stats

scipy.stats.shapiro(residual)
```

```python
fitted = naive
residual = real - fitted

res_df = pd.DataFrame({
    'fitted':fitted,
    'residual':residual
})

sns.regplot(x="fitted", y="residual", lowess=True, data=res_df, line_kws={'color':'red'})
plt.plot([50, 300], [0,0], '--', color='grey')
```

```python
scipy.stats.shapiro(residual)
```
